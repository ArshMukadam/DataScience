```python
import numpy as np

moods = ["sad","tired","bored","chill","excited","angry"]

intensity = np.array([[1],[2],[3],[4],[5],[6]])

movies = {
    1: ["Comedy","Drama","Animated-Film"],
    2: ["Heist","Adventure","Sport"],
    3: ["Sci-fi","Action-Adventure","Mystery"],
    4: ["Documentary","Romance","Horror"],
    5: ["Thriller","Suspense","Historical-Films"],
    6: ["War","Superhero","Action-Comedy"]}

print("Mood Type: angry, sad, tired, bored, chill, excited.")
user_mood = input("Enter your mood: ").lower()

if user_mood in moods:
    index = moods.index(user_mood)
    level =intensity[index][0]
    suggestion = np.random.choice(movies[level])

    print("\n---Movie Category Suggestion---")
    print(f"Mood Detected: {user_mood}")
    print(f"Intensity    : {level}")
    print(f"Category     : {suggestion}")
else:
    print("Unknown Mood! Please try: angry, sad, tired, bored, chill, excited.")
    
```
