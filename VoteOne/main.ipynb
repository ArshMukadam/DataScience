```python
import time
from datetime import datetime

def read_lines(file):
    with open(file, "r")as f:
        return [line.strip() for line in f if line.strip()]

def write_lines(file,lines):
    with open(file, "w")as f:
        for line in lines:
            f.write(line + "\n")
            
def append_line(file,line):
    with open(file, "a")as f:
        f.write(line + "\n")

def load_voters():
    voters = {}
    for line in read_lines("voters.txt"):
        vid,name,age,booth,voted = line.strip().split(",")
        voters[vid] ={
            "id": vid,
            "name": name,
            "age": int(age),
            "booth": booth,
            "voted":voted == "True"
        }
        return voters
        
def save_voters(voters):
    lines=[]
    for v in voters.values():
        lines.append(f"{v['id']},{v['name']},{v['age']},{v['booth']},{v['voted']}")
    write_lines("voters.txt",lines)
    
def load_candidates():
    cands = {}
    for line in read_lines("candidates.txt"):
        cid,name,party,booth,votes = line.split(",")
        cands[cid] ={
            "id": cid,
            "name": name,
            "party": party,
            "booth": booth,
            "votes":int(votes)
        }
        return cands

def save_candidates(cands):
    lines=[]
    for c in cands.values():
        lines.append(f"{c['id']},{c['name']},{c['party']},{c['booth']},{c['votes']}")
    write_lines("candidates.txt",lines)        
```

```python
def cast_vote(voter_id,candidate_id):
    voters = load_voters()
    candidates = load_candidates()

    if voter_id not in voters:
        print(" Voter not found")
        return

    if candidate_id not in candidates:
        print(" Candidate not found")
        return

    voter = voters[voter_id]

    if voter["voted"] == True:
        print(" You have already voted")
        return

    candidates[candidate_id]["votes"] +=1
    voter["voted"] = True

    save_candidates(candidates)
    save_voters(voters)

    print(" Vote Cast Successfully ! ")
```
